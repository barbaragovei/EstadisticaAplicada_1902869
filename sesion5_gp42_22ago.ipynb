### 1. Librerias

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. Datos

```python
personas = ['Jorge','Raul','Ana','Lissa']
edades = [27,16,21,30]
dic1={
    #'Key':'Value'
    'Municipio':['Escobedo','Apodaca','Guadalupe','Santiago'],
    'Nacimientos':[123,345,278,321],
    'Muertes':[23,17,56,32],
    'Matrimonios':[76,45,89,12]
}
datos_mun = pd.DataFrame(dic1)
datos_mun
```

### 3. Graficas con MatPlotLib

```python
plt.bar(personas,edades)
plt.title('Edades')
plt.xlabel('Nombres')
plt.ylabel('Edad en aÃ±os')
plt.show() # Para quitar el textito je
```

```python
plt.bar(datos_mun['Municipio'],datos_mun['Matrimonios'])
plt.title('Matrimonios en NL')
plt.xlabel('Municipio')
plt.ylabel('Num. de matrimonios')
plt.show() # Para quitar el textito je
```

```python
# Para hacer 2 graficas juntas
# fig es el nombre de pizarron o lienzo
fig, (graf1,graf2) = plt.subplots(2,1) #2 filas,1 columna
graf1.bar(personas,edades)
graf2.bar(datos_mun['Municipio'],datos_mun['Matrimonios'])
plt.show()
```

### GrÃ¡ficas con Pandas
El graficador de pandas esta construido sobre matplotlib al igual que seaborn

dataframe.plot.tipografica()

```python
# Grafico de barras
# Pone en el eje x a los indices
datos_mun.plot.bar() # Graficando con pandas
plt.show()
```

```python
datos_mun = datos_mun.set_index('Municipio')
datos_mun['Matrimonios'].plot.bar() # Graficando con pandas
plt.title('Matrimonios en NL')
plt.xlabel('Municipio')
plt.ylabel('# de Matrimonios')
plt.show()
```

```python
# Grafico de barras apiladas
datos_mun.plot.bar(stacked=True)
plt.show()
```

```python
## Grafico de lineas
datos_mun.plot.line()
plt.show()
```

```python
# Grafico de areas
datos_mun.plot.area()
plt.show()
```

```python
## Histografa
datos_mun['Nacimientos'].plot.hist()
plt.xlabel('Municipio')
plt.ylabel('# de Nacimientos')
plt.show()
```
